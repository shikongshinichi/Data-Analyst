##### Người mắc bệnh ung thư cần tránh những gì
##### Người mắc bệnh ung thư thường có những triệu chứng gì rõ rệt nhất
##### Những nguyên nhân dẫn đến việc mắc bệnh ung thư, nguyên nhân nào là cao nhất?
##### Sự thay đổi trong triệu chứng bệnh/ cách sinh hoạt khi ung thư trở nặng

```python
df1 = pd.read_csv('cancer patient_feature.csv')
df2 = pd.read_csv('cancer patient_symptom.csv')
```

```python
df1.columns
```

```python
df2.columns
```

```python
df2 = df2.drop(columns="Patient Id")
df_mean = df2.groupby('Level').mean().reset_index()

df_melted = df_mean.melt(id_vars='Level', var_name='Symptom', value_name='Mean Value')
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_melted, x='Symptom', y='Mean Value', hue='Level', marker='o')

plt.title('Biểu đồ đường biểu diễn giá trị trung bình của các triệu chứng theo Level')
plt.xticks(rotation=90)
plt.xlabel('Triệu chứng')
plt.ylabel('Giá trị trung bình')
plt.legend(title='Level')
plt.tight_layout()
plt.xticks(rotation=45)

plt.show()
```

#### Nhận xét: 
##### Từ mức độ nhẹ lên trung bình, các chỉ số hầu như tăng đồng đều, tỉ lệ bệnh nhân bị biến dạng móng tay chân (Clubbing of Finger Nails) tăng nhiều hơn so với các dấu hiệu khác
##### Từ mức độ trung bình đến nặng, tỉ lệ bệnh nhân ho ra máu (Coughing of blood) là tăng đột biến, tỉ lệ giảm cân (Weight loss), ngáy ngủ (Snoring) và tỉ lệ bệnh nhân khó nuốt khi ăn uống (Swallowing difficutty) gần như không thay đổi, tỉ lệ thở khò khè (Wheezing) và biến dị móng tay chân (Clubbing of Finger Nails) giảm xuống so với mức độ trước

```python
needed_columns = ['Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Obesity', 'Smoking',
       'Passive Smoker']
sum_values = df1[needed_columns].sum()
```

```python
plt.figure(figsize=(10,10))
plt.pie(sum_values, labels=needed_columns, autopct='%1.1f%%', startangle=140)
plt.title('Tỉ lệ nguyên nhân mắc ung thư')
plt.show()
```

#### Nhận xét:
##### Nhìn chung tỉ lệ các nguyên nhân dẫn đến ung thư phổi là tương đối cân bằng, có thể thấy ung thư phổi xuất hiện do nhiều nguyên nhân, không có nguyên nhân nào là chủ yếu, nguyên nhân cao nhất là do Dị ứng bụi (Dust Allergy)
##### Vì vậy, người dân nên có các biện pháp phòng tránh tiếp xúc với bụi như đeo khẩu trang, tránh những công trường, nơi ô nhiêm không khí do khói bụi, bên cạnh đó tránh đồ uống có cồn, hút thuốc, và tránh những người hút thuốc (ta có thể thấy hút thuốc bị động chiếm tỉ lệ cao hơn so với tự hút thuốc)

```python
plt.figure(figsize=(10,4))
plt.hist(df1['Age'], bins=30)
plt.title("Phân bố độ tuổi của bệnh nhân")
plt.xlabel('Độ tuổi')
plt.ylabel('Số lượng')
plt.show()
```

#### Nhận xét:
##### Theo khảo sát, độ tuổi trung bình của bệnh nhân ung thư phổi rơi vào khoảng từ 30 đến 45 tuổi
##### Tỉ lệ độ tuổi này chiếm đông đảo có thể xuất phát từ việc đây là độ tuổi còn đang đi làm, phải ra đường gặp nhiều tác nhân gây ung thư phổi, ngoài ra có thể coi như từ độ tuổi 55 đổ lên không tham gia nhiều vào bài khảo sát vì nguyên nhân tiếp cận công nghệ, đối tượng đưa ra khảo sát

```python
#Giả sử df là dataset đã merge
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 45)]
```

```python
needed_columns = ['Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Obesity', 'Smoking',
       'Passive Smoker']
sum_values = filtered_df[needed_columns].sum()
```

```python
plt.figure(figsize=(10,10))
plt.pie(sum_values, labels=needed_columns, autopct='%1.1f%%', startangle=140)
plt.title("Nguyen nhan mac ung thu phoi o do tuoi 30-45")
plt.show()
```

#### Nhận xét:
##### Có thể thấy nguyên nhân dẫn đến ung thư phổi theo khảo sát cao nhất vấn là do dị ứng bụi (Dust Allergy), vậy nên nguyên nhân không có sự khác biệt giữa phân bố độ tuổi chiếm tỉ lệ mắc ung thư phổi cao và tổng thể toàn bộ độ tuổi
