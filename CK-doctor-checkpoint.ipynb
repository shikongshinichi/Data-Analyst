## Tần suất mắc bệnh phổi mãn tính thay đổi như thế nào theo mức độ ô nhiễm không khí?

```python
# Tính tần suất mắc bệnh phổi mãn tính theo mức độ ô nhiễm không khí
air_pollution_groups = df.groupby('Air Pollution')['chronic Lung Disease'].sum()

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.bar(air_pollution_groups.index, air_pollution_groups.values, color='blue')
plt.xlabel('Mức độ Ô nhiễm Không khí')
plt.ylabel('Tần suất Mắc Bệnh Phổi Mãn Tính')
plt.title('Tần suất mắc bệnh phổi mãn tính theo mức độ ô nhiễm không khí')
plt.xticks(air_pollution_groups.index)
plt.show()
```

### Nhìn chung tần suất mắc bệnh phổi mãn tính không đồng đều so với các mức độ ô nhiễm không khí.
### Ta thấy ở mức độ ô nhiễm không khí ở mức 5 nhưng tần suất mắc bệnh lại thấp nhất, còn ở mức 6 thì lại có tần suất mắc bệnh cao nhất.
### Như vậy ta có thể thấy, ngoài mức độ ô nhiễm không khí còn có nhiều yếu tố khác ảnh hưởng tới tần suất mắc bệnh phổi mãn tính này
### Nhưng mọi người cũng không nên chủ quan, luôn phải có những biện pháp bảo vệ cơ thể khỏi ô nhiễm môi trường như: đeo khẩu trang, dọn dẹp vệ sinh quanh khu ở, trồng thêm nhiều cây xanh,...

## Tỷ lệ mắc bệnh phổi mãn tính ở những người tiếp xúc với khói thuốc như thế nào so với người hút thuốc?

```python
# Tính tỷ lệ mắc bệnh phổi mãn tính ở mỗi mức độ tiếp xúc với khói thuốc
passive_smoker_levels = df.groupby('Passive Smoker')['chronic Lung Disease'].mean()

# Tính tỷ lệ mắc bệnh phổi mãn tính ở mỗi mức độ hút thuốc
smoking_levels = df.groupby('Smoking')['chronic Lung Disease'].mean()

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))

# Biểu đồ cho mức độ tiếp xúc với khói thuốc
plt.subplot(1, 2, 1)
passive_smoker_levels.plot(kind='bar', color='skyblue')
plt.title('Chronic Lung Disease by Passive Smoker Level')
plt.xlabel('Passive Smoker Level')
plt.ylabel('Chronic Lung Disease Rate')

# Biểu đồ cho mức độ hút thuốc
plt.subplot(1, 2, 2)
smoking_levels.plot(kind='bar', color='lightgreen')
plt.title('Chronic Lung Disease by Smoking Level')
plt.xlabel('Smoking Level')
plt.ylabel('Chronic Lung Disease Rate')

plt.tight_layout()
plt.show()
```

### Nhìn qua 2 biểu đồ ta thấy giữa người tiếp xúc với khói thuốc và người hút thuốc đều có tỷ lệ mắc bệnh phổi gần giống như nhau.
### Tuy có sự chênh lệch giữa 2 đối tượng nhưng không đáng kể
### Vì vậy để giảm tỷ lệ mắc bệnh phổi, hãy thực hiện các biện pháp cai nghiện hút thuốc với những người đang hút thuốc như: nhai kẹo thay vì hút thuốc, sử dụng các sản phảm giảm mức độ thèm hút thuốc,... và khi hút thuốc hãy tránh xa những người bên cạnh không có hút thuốc đặc biệt là người già, phụ nữ mang thai, trẻ em,...

## Sự khác nhau của nguy cơ mắc bệnh phổi mãn tính khi sử dụng rượu bia không?

```python
# Tính tỷ lệ mắc bệnh phổi mãn tính cho mỗi mức độ sử dụng rượu bia
alcohol_levels = df.groupby('Alcohol use')['chronic Lung Disease'].mean()

# Tạo biểu đồ
plt.figure(figsize=(8, 6))
plt.bar(alcohol_levels.index, alcohol_levels.values)
plt.title('Chronic Lung Disease by Alcohol Use Level')
plt.xlabel('Alcohol Use Level')
plt.ylabel('Mean Chronic Lung Disease')
plt.show()
```

### Ta thấy qua biểu đồ trên mức độ mắc bệnh tăng dần theo mức độ sử dụng rượu bia
### Nhưng có sự khác biệt nhỏ là ở mức độ 7 của sử dụng rượu bia lại có tỷ lệ mắc bệnh cao nhất, cao hơn cả mức độ 8 của việc sử dụng rượu bia, có lẽ có những yếu tố khác đã tác động tới làm ảnh hưởng tỷ lệ ở mức độ 7 và mức độ 8 của việc sử dụng rượu bia, có thể là cân nặng, chế độ ăn uống, độ tuổi,...
### Để giảm tỷ lệ mắc bệnh ở nguyên nhân sử dụng rượu bia này chúng ta cần áp dụng pháp luật nghiêm hơn nữa để giảm tình trạng sử dụng rượu bia, nhằm tránh tỷ lệ mắc bệnh cũng như tai nạn giao thông và một số vấn đề khác

## Mức độ khó khăn khi nuốt có xu hướng thay đổi như thế nào dựa trên việc sử dụng rượu 

```python
sns.barplot(y='Alcohol use', x ='Swallowing Difficulty', data=df)
```

### Sự khó khăn khi nuốt có xu hướng không đồng đều khi sử dụng rượu bia, cụ thể như:
     Ở khoảng giữa mức độ 3 và 4 của việc sử dụng rượu bia có sự khó khăn khi nuốt thấp nhất
     Ở mức độ sử dụng rượu bia là 7 thì sự khó khăn khi nuốt lại cao nhất
     Điều đặc biệt là sử dụng rượu bia ở mức độ 5 thì sự khó khăn khi nuốt ở mức độ 1
### Khi nguy cơ mắc bệnh phổi để giảm sự khó chịu khi nuốt chúng ta cần hạn chế sử dụng các chất cồn như rượu, bia. Khi bị bệnh chúng ta phải tránh thực hiện các việc ảnh hưởng xấu tới tình trạng bệnh của bản thân

## Tần suất ngủ ngáy có khác nhau ở những bệnh nhân mắc bệnh phổi mãn tính và không mắc không?

```python
# Vẽ biểu đồ tần suất ngủ ngáy
plt.figure(figsize=(8, 6))
sns.barplot(x='chronic Lung Disease', y='Snoring', data=df)
plt.title('Tần suất ngủ ngáy ở bệnh nhân mắc bệnh phổi mãn tính và không mắc')
plt.xlabel('Bệnh phổi mãn tính')
plt.ylabel('Tần suất ngủ ngáy')
plt.show()
```

### Qua biểu đồ, ta thấy khi mới có triệu chứng mắc bệnh phổi tần suất ngủ ngáy đã xuất hiện khoảng 1.5
### Tới giai đoạn mắc bệnh ở mức độ 2 tần suất ngủ ngáy tăng lên 1 cách rõ rệt khoảng 3.5
### Trong quá trình mắc bệnh thường tần suất ngủ ngáy tăng giảm khoảng 2.0 tới 3.0
### Ở mức độ bệnh phổ mãn tính cao nhất thì tần suất ngủ ngáy cũng cao nhất là trên 3.5
### Do vậy, thông qua triệu chứng ngủ ngáy ta có thể xác định mức độ bệnh phổi cũng mình để có thể kịp thời phát hiện các bệnh liên quan cụ thể là bệnh phổi và khám chữa đúng lúc.
