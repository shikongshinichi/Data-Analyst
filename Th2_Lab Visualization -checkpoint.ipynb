```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data
df = pd.read_csv(r"D:\Documents\3nd_year\Semester2\Thematic 2\Data\job_survey_clean.csv")
df.head()
```

```python
df.columns.tolist()
```

#### 1. Số lượng người dùng và mức lương từng ngôn ngữ


```python
# get data
df_language = df.groupby('language').agg({'id':'count', 'salary':'mean'}).reset_index()
df_language
```

```python
# draw bar chart
df_language.plot(kind='bar', x='language')
plt.title('Number of user and salary by language')
plt.grid()
plt.xticks(rotation=30)
```

Observation: 
- python has the highest number of users
- java has the lowest number of users
- java  language has the highest average salary
- c/c++ language has the lowest average salary

#### 2. Sự phổ biến của từng nhóm Job - Sử dụng pie


```python
# get data
df_jobs = df.groupby('curent_role').count()['id']
df_jobs
```

```python
# draw pie chart
plt.figure(figsize=(14, 6))
df_jobs.plot(kind='pie', autopct='%.1f%%', startangle=90)
plt.title('Current role Distribution')
```

Observation:
- Data Analyst is the most popular job.
- Database Architect has the lowset amount.

#### 3. Mức lương theo từng nhóm Job


```python
# get data
df_salary = df.groupby('curent_role')['salary'].mean()
df_salary
```

```python
# draw bar chart
df_salary.plot(kind='bar')
plt.title('Salary by job')
plt.grid()
plt.xticks(rotation=30)
```

Observation:
- Data Scientist has the highest average salary
- Student/Looking/None has the lowest average salary

#### 4. Mức độ Happy với từng Job - Sử dụng sctter


```python
# get data
df_happy = df.groupby('curent_role')['happy_salary'].mean()
df_happy
```

```python
# draw scatter chart
plt.scatter(df_happy.index, df_happy.values)
plt.xticks(rotation=30)
```

Observation:
- Data Scientist has the highest happy salary score
- Student/Looking/None has the lowest happy salary score

#### 5. Mức lương trung bình theo từng nhóm job của từng ngôn ngữ

```python
# get data
df_language_salary = df.groupby(['language', 'curent_role']).agg({'salary':'mean'}).reset_index()
df_language_salary
```

```python
# draw bar chart
plt.figure(figsize=(14, 7))
sns.barplot(data=df_language_salary, x='language', y='salary', hue='curent_role')
plt.title('Salary by Job and Language')
plt.legend(loc=2)
plt.grid()
plt.xticks(rotation=30)
```

Observation:
- Not all kinds of language in this servey have all kind of job.
- Python is the language has all kind of jobs.
- The Job has highest salary in each row are c/c++: Other, Java: Other, javascript: Data Engineer, other: Data Engineer, Python: Data Scientist, R: Data Scientist.
- No another job use Java but Other job.
