```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.info-bets.com/stats-form.aspx?l=2333"
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find_all('table')[0]
```

```python
table = soup.find_all('table')[0]
```

```python
titles =  soup.find_all('th')
```

```python
 tb_titles = [title.text.strip() for title in titles]
```

```python
print(tb_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = tb_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
print(column_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    indi_row = [data.text.strip() for data in row_data]
    
    print(indi_row)
    length = len(df)
    print(length)
    df.loc[length] = indi_row
```

```python
df.to_csv(r'', index = False)
```
