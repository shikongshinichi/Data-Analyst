## 3. Clean data
#### Implementer: Nguyễn Thị Tiểu Tiên

Clean the data to solve the many issues corresponding to data quality and tidiness found in the assessing step. 

Make copies of the datasets 

```python
df_feature_clean = df_feature.copy()
df_symptom_clean = df_symptom.copy()
```

### Dataset 1: Patient's feature

Drop NULL values:

```python
df_feature_clean.dropna(inplace=True)
```

```python
# Validate the cleaning was successful
df_feature_clean.isnull().sum()
```

Convert data to suitable data type to save memory:

```python
# function gets a array of column names and convert these columns to 'int64'
def convert_to_int64(df_clean, arr):
    df = df_clean
    for i in arr:
        df[i] = df[i].astype('int64')
    return df
```

```python
# choose array of columns that should be changed data type to 'int64'
arr = ['Air Pollution', 'Alcohol use', 'Dust Allergy', 'OccuPational Hazards', 'chronic Lung Disease', 'Balanced Diet']
# call convert_to_int64 function
df_feature_clean = convert_to_int64(df_feature_clean, arr)
```

```python
# convert 'Gender' to 'bool' data type
df_feature_clean['Gender'] = (df_feature_clean['Gender'] == 1)
```

```python
# Validate the cleaning was successful
df_feature_clean.dtypes
```

### Dataset 2: Patient's symptom

Drop NULL values:

```python
df_symptom_clean.dropna(inplace=True)
```

```python
# Validate the cleaning was successful
df_symptom_clean.isnull().sum()
```

Convert to suitable data types:

```python
# convert to 'int64'
arr = ['Coughing of Blood', 'Weight Loss', 'Wheezing', 'Swallowing Difficulty']
df_symptom_clean = convert_to_int64(df_symptom_clean, arr)
```

```python
# Validate the cleaning was successful
df_symptom_clean.dtypes
```

```python
# convert 'Level' to 'int64'
def convert_level(level):
    if level == 'Low': 
        return 1
    if level == 'Medium':
        return 2
    else:
        return 3

# Use above funtion to change to suitable level number and convert to 'int64' data type
df_symptom_clean['Level'] = df_symptom_clean['Level'].apply(convert_level).astype('int64')
```

```python
# Validate the cleaning was successful
df_symptom_clean['Level'].unique()
```

### Combine datasets

```python
# merge 2 dataset based on 'Patient Id'
df_merged = pd.merge(df_feature_clean, df_symptom_clean, on='Patient Id', how='inner')
```

```python
# function to set label for chart
def set_plot_labels(title, xlabel, ylabel):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

### Question 1: How does the incidence of chronic lung disease vary with air pollution levels?
#### Implementer: Nguyễn Kim Ngân

```python
# get data group by 'Air Pollution' and sum values 'chronic Lung Disease' in each group
air_pollution_groups = df.groupby('Air Pollution')['chronic Lung Disease'].sum()

# draw bar chart
plt.figure(figsize=(10, 6))
plt.bar(air_pollution_groups.index, air_pollution_groups.values, color='blue')
set_plot_labels("Chronic lung disease vary with air pollution levels", "Air Pollution Level", "Chronic Lung Disease Level")
plt.xticks(air_pollution_groups.index)
plt.show()
```

#### Comment:
- In general, the incidence of chronic lung cancer is not uniform compared to air pollution levels.
- Ex: At air pollution level 5, the cencer frequency is the lowest, while at level 6,  the cencer frequency is the highest.

=>  In addition to the level air pollution, there are many other factors that affect the frequency of choric lung disease.
#### Recommend: 
- Protecting enviroment also help people decreasing the frequency of choric lung disease.
- Shouls always take methods to protect the body from enviromental pollution, such as: wearing masks, cleaning around the living area, plating more trees,...

### Question 2: Compare the rate of chronic lung disease between passive smoking people and smokers.
#### Implementer: Nguyễn Kim Ngân

```python
# The incidence of chronic lung disease at each level of passive smoke
passive_smoker_levels = df.groupby('Passive Smoker')['chronic Lung Disease'].mean()

# The incidence of chronic lung disease at each level of smoking
smoking_levels = df.groupby('Smoking')['chronic Lung Disease'].mean()

# draw chart
plt.figure(figsize=(10, 6))

# Chart for passive smoke
plt.subplot(1, 2, 1)
passive_smoker_levels.plot(kind='bar', color='skyblue')
set_plot_labels('Chronic Lung Disease by Passive Smoker Level', 'Passive Smoker Level', 'Chronic Lung Disease Rate')

# Chart for smoking
plt.subplot(1, 2, 2)
smoking_levels.plot(kind='bar', color='lightgreen')
# set label by function
set_plot_labels('Chronic Lung Disease by Smoking Level', 'Smoking Level', 'Chronic Lung Disease Rate')
# adjusts the spacing between subplots
plt.tight_layout()
plt.show()
```

#### Comment:
- Through 2 charts, passive smoking people and smokers have alomost the same rate of lung cancer (some different but not significant).
#### Recommend:
- To reduce the incidence of lung cancer, please stop smoking.
- When smoking, stay away from around peple who not smoking, especially the elderly, pregnant woman, children.

### Question 7: Which disease symptoms will change when cancer gets worse?
#### Implementer: Đinh Thị Lan Hương

```python
# get data about symptoms by some colums below
df_symptom = df[['Chest Pain',
 'Coughing of Blood',
 'Fatigue',
 'Weight Loss',
 'Shortness of Breath',
 'Wheezing',
 'Swallowing Difficulty',
 'Clubbing of Finger Nails',
 'Frequent Cold',
 'Dry Cough',
 'Snoring',
 'Level']]
df_mean = df_symptom.groupby('Level').mean().reset_index()
df_melted = df_mean.melt(id_vars='Level', var_name='Symptom', value_name='Mean Value')
# draw line chart
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_melted, x='Symptom', y='Mean Value', hue='Level', marker='o')
# set label by function
set_plot_labels('Variation of symptoms according to severity', 'Symptom', 'Avarage Value')
plt.xticks(rotation=90)
plt.legend(title='Level')
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```

#### Comment:

- From mild to moderate severity, the indexes increase almost uniformly, the proportion of patients with deformed fingernails (Clubbing of Finger Nails) increases more than other signs.
- From moderate to severe, the rate of patients coughing up blood (Coughing of blood) increases dramatically, the rate of weight loss (Weight loss), snoring (Snoring) and the rate of patients having difficulty swallowing when eating ( Swallowing difficutty) remained almost unchanged, the rate of wheezing and Clubbing of Finger Nails decreased compared to previous levels.

#### Recommend:
- If the patient shows signs of coughing up blood and feels increased fatigue, this may be a sign that the patient's lung cancer is getting worse.

### Question 8: What are the most obvious characteristics that cancer patient often have?
#### Implementer: Đinh Thị Lan Hương

```python
# get data about characteristics of patient from below column list
needed_columns = ['Air Pollution', 'Alcohol use',
       'Dust Allergy', 'OccuPational Hazards', 'Genetic Risk',
       'chronic Lung Disease', 'Obesity', 'Smoking',
       'Passive Smoker']
sum_values = df[needed_columns].sum()
# draw pie chart
plt.figure(figsize=(10,10))
plt.pie(sum_values, labels=needed_columns, autopct='%1.1f%%', startangle=140)
# set title for chart 
plt.title('Ratio of lung cancer causes')
plt.show()
```

#### Recommend:

-  People should take measures to prevent exposure to dust such as wearing masks, avoiding construction sites and places with air pollution due to dust and smoke, in addition to avoiding alcoholic beverages, smoking, and avoiding smokers (we can see that passive smoking accounts for a higher rate than self-smoking).
