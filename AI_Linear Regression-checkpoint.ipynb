```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# declare/ define X (input), y (target)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
# T: each row eah data point, each col an attribute - struct for input to funcs
```

```python
# visualize data
plt.plot(X, y, 'ro')
plt.axis([140, 190, 45, 75])
plt.show()
```

Observation: seem like on 1 straight line
=> LR model will give good result 

```python
# y = w0 + w1*x0
# create X
bias = np.ones((X.shape[0], 1))
Xbar = np.concatenate((bias, X), axis=1)
# normal regession -> w vector
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print(w)
# 
w_0 = w[0][0]
w_1 = w[1][0]
x_0 = np.linspace(145, 185, 2) #2: num of samples
y_0 = w_0 + w_1*x_0
```

```python
plt.plot(X.T, y.T, 'ro')
plt.plot(x_0, y_0)
plt.show()
```
