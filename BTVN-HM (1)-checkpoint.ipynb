# Xây dựng mô hình hồi quy tuyến tính đơn biến

```python
#Kiểm tra sự tương qua giữa 2 biến bằng biểu đồ
plt.scatter(df['age'], df['totChol'])
plt.xlabel('age')
plt.ylabel('totChol')
plt.grid(True)
plt.show()
```

```python
# Chọn biến độc lập và phụ thuộc
X = df[['age']]  # Biến độc lập
y = df['totChol']    # Biến phụ thuộc

# Thêm cột intercept vào dữ liệu X
X = sm.add_constant(X)

# Tính toán mô hình hồi quy tuyến tính
model = sm.OLS(y, X).fit()

# Lấy các hệ số alpha và beta
alpha = model.params['const']
beta = model.params['age']

print("Alpha (Intercept):", alpha)
print("Beta (Slope):", beta)
```

```python
# Vẽ scatter plot của dữ liệu
plt.scatter(df['age'], df['totChol'], color='blue', label='Data')

# Vẽ đường hồi quy
plt.plot(df['age'], alpha + beta * df['age'], color='red', label='Regression Line')

# Đặt nhãn và tiêu đề
plt.xlabel('age')
plt.ylabel('totChol')
plt.title('Linear Regression')

# Thêm chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

# Xây dựng mô hình hồi quy tuyến tính đa biến

```python
X = df[['age', 'BMI']] #biến độc lập
y = df['totChol'] #biến phụ thuộc
X = sm.add_constant(X) # Thêm cột intercept vào dữ liệu X
model = sm.OLS(y, X).fit() # Tính toán mô hình hồi quy tuyến tính
# Lấy các hệ số alpha và beta1, beta2
alpha = model.params['const']
beta1 = model.params['age']
beta2 = model.params['BMI']

print("A1", alpha)
print("Beta1 ", beta1)
print("Beta2 ", beta2)
```

```python
# Tạo một lưới dữ liệu cho biểu đồ
x1_grid, x2_grid = np.meshgrid(np.linspace(0, 10, 10), np.linspace(0, 5, 10))
Z = model.predict(np.cumnol_stack((x1_grid.ravel(), x2_grid.ravel())))
Z = Z.reshape(x1_grid.shape)

# Vẽ biểu đồ 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['age'], df['BMI'], df['totChol'], color='blue', label='Data points')
ax.plot_surface(x1_grid, x2_grid, Z, alpha=0.5, cmap='viridis', label='Linear regression surface')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.legend()
```
