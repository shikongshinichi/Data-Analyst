```python
# chuyen doi string sang datetine
df.date_taken = pd.to_datetime(df.date_taken)
```

```python
df[['happy_salary', 'work_life_balance']] = df[['happy_salary', 'work_life_balance']].astype('float')
```

```python
# thay cac o rong thanh null (NA)
# tu null moi co the chuyen thanh string, number
df.replace(r'^s*$', np.nan)
df
```

```python
# thay cac gia tri Other... = Other
df['curent_role'] = df['curent_role'].str.replace('Other.*', 'Other', regex=True)
df['curent_role'].unique()
```

```python
# kiem tra cot salary
df['salary'].unique()
```

#### Tách cột salary thành 2 cột khác nhau

```python
df[['sal_min', 'sal_max']] = df['salary'].str.split('-', n=2, expand=True)
df.head()
```

```python
# chi lay so, ko lay chu K trong 2 cot salary
df['sal_min'] = df['sal_min'].str.extract('(\d+)')
df['sal_max'] = df['sal_max'].str.extract('(\d+)')
df.head()
```

```python
df['sal_max'].unique()
```

```python
# truong hop sal_max = null (255K+) -> max = min
df.loc[df['sal_max'].isna(), 'sal_max'] = df['sal_min']
df['sal_max'].unique()
```

```python
# ep kieu cho salary
df[['sal_min', 'sal_max']] = df[['sal_min', 'sal_max']].astype(float)
df[['sal_min', 'sal_max']].info()
```

```python
# tinh salary tb 
df['salary'] = (df['sal_min'] + df['sal_max'])/2
df['salary']
```

```python
# xoa 2 cot sub_sal
df.drop(['sal_min', 'sal_max'], axis=1, inplace=True)
df.columns
```

```python
# thay the cac Other... = Other
df = df.replace('Other.*', 'Other', regex=True)
```

```python
df['industry'].unique()
```

```python
# thay the tuoi>100 = tuoi tb
df.loc[df['age']>100, 'age'] = df['age'].mean()
df['age'].max()
```
