```python
# replace missing value by mean value
df3 = df
mean = df['Age'].mean()
df3['Age'].fillna(mean, inplace=True)
df3
```

```python
# check outliers value of 'Age' by boxplot chart
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df3['Age'])
```

```python
# tìm giá trị q1, q3
q1 = df3['Age'].quantile(0.25)
print(q1)
q3 = df3['Age'].quantile(0.75)
q3

# import numpy as np
# q1, q3 = np.percentile(df3['Age'], [25, 75])
# print(q1)
# print(q3)
```

```python
# Calculate the interquartile range
iqr = q3 - q1
# Calculate the lower and upper bounds
#lower_bound = q1 - (1.5 * iqr)
#upper_bound = q3 + (1.5 * iqr)

lower_bound = q1 - (iqr)
upper_bound = q3 + (iqr)

# Drop the outliers
clean_data = df3[(df3['Age'] >= lower_bound) & (df3['Age'] <= upper_bound)]
```

```python
# create a new dataset not contain outlier values
# Vẽ biểu đồ Box plot cho biến Age
plt.figure(figsize=(8, 6))
plt.boxplot(clean_data['Age'], vert=False)
plt.xlabel('Age')
plt.title('Box Plot for Variable Age')
plt.show()
```

```python
# Thống kê số lượng người sống/chết theo giới tính
survived_by_sex = df3.groupby(['Sex', 'Survived']).size().unstack()
print(survived_by_sex)
survived_by_sex.plot(kind='bar', stacked=True)
plt.title('Sống/Chết theo Giới tính')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng')
plt.xticks(rotation=0) # Xoay nhãn trục x nếu cần thiết
plt.legend(title='Survived', labels=['Chết', 'Sống'])
plt.show()
```

```python
# chuyển thành cột song song
survived_by_sex.plot(kind='bar', stacked=False)
plt.title('Sống/Chết theo Giới tính')
plt.xlabel('Giới tính')
plt.ylabel('Số lượng')
plt.xticks(rotation=0) # Xoay nhãn trục x nếu cần thiết
plt.legend(title='Survived', labels=['Chết', 'Sống'])
plt.show()
```

```python
# Thống kê theo nhân xưng (Mr, Miss, Master,Mrs,vv...) của biến Name và vẽ biểu đồ minh họa

# Tách tên và họ từ cột "Name"
df3['Last Name'] = df3['Name'].apply(lambda x: x.split(',')[0])
df3['Title'] = df3['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())

# Thống kê số lượng hành khách theo từng loại tiêu đề (Title)
title_counts = df3['Title'].value_counts()

# Vẽ biểu đồ minh họa kết quả
plt.figure(figsize=(10, 6))
title_counts.plot(kind='bar')
plt.title('Passenger Counts by Title')
plt.xlabel('Title')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=45)
plt.show()
```

## Preprocessing

```python
# missing vslue
df.isnull().mean()
```

```python
# with above percent of missing value of each column, decide:
- drop the column ADDRESSLINE2: percent high and string dtype
- STATE: replace by frecency
- POSTALCODE: crop row - percent is low
- TERRITORY: replace by frecency
```

```python
# drop the column ADDRESSLINE2
df2 = df.drop(columns=['ADDRESSLINE2'])
df2.isnull().mean()
```

```python
# STATE: replace by frecency
mode_state_value = df['STATE'].mode()[0]
mode_state_value
df2['STATE'] = df['STATE'].fillna(mode_state_value)
df2.isnull().mean()
```

```python
# POSTALCODE: crop row - percent is low
df2 = df2.dropna(subset=['POSTALCODE'], axis=0)
df2.isnull().mean()
```

```python
# TERRITORY: replace by frecency
frecency = df['TERRITORY'].mode()[0]
frecency
df2['TERRITORY'] = df2['TERRITORY'].fillna(frecency)
df2.isnull().sum()
```
