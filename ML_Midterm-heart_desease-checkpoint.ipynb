# Xử lý outliers

### outliers của cột cigsPerDay

```python
sns.boxplot(x=df['cigsPerDay'])
```

```python
# Xử lý outliers (ví dụ: có thể loại bỏ các điểm nằm ngoài 1.5*IQR) - tứ phân vị
Q1 = df['cigsPerDay'].quantile(0.25)
Q3 = df['cigsPerDay'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['cigsPerDay'] > lower_bound) & (df['cigsPerDay'] < upper_bound)]
#chạy lại biểu đồ sau khi xử lý outliers
sns.boxplot(x=df['cigsPerDay'])
```

# Xây dựng mô hình hồi quy tuyến tính đơn biến

```python
#Kiểm tra sự tương qua giữa 2 biến bằng biểu đồ
plt.scatter(df_clean['age'], df_clean['totChol'])
plt.xlabel('age')
plt.ylabel('totChol')
plt.grid(True)
plt.show()
```

```python
# Chọn biến độc lập và phụ thuộc
X = df_clean[['age']]  # Biến độc lập
y = df_clean['totChol']    # Biến phụ thuộc

# Thêm cột intercept vào dữ liệu X
X = sm.add_constant(X)

# Tính toán mô hình hồi quy tuyến tính
model = sm.OLS(y, X).fit()

# Lấy các hệ số alpha và beta
alpha = model.params['const']
beta = model.params['age']

print("Alpha (Intercept):", alpha)
print("Beta (Slope):", beta)
```

```python
# Vẽ scatter plot của dữ liệu
plt.scatter(df_clean['age'], df_clean['totChol'], color='blue', label='Data')

# Vẽ đường hồi quy
plt.plot(df_clean['age'], alpha + beta * df_clean['age'], color='red', label='Regression Line')

# Đặt nhãn và tiêu đề
plt.xlabel('age')
plt.ylabel('totChol')
plt.title('Linear Regression')

# Thêm chú thích
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

# Xây dựng mô hình hồi quy tuyến tính đa biến

```python
X = df_clean[['age', 'BMI']] # biến độc lập
y = df_clean['totChol'] # biến phụ thuộc
X = sm.add_constant(X) # Thêm cột intercept vào dữ liệu X
model = sm.OLS(y, X).fit() # Tính toán mô hình hồi quy tuyến tính
# Lấy các hệ số alpha và beta1, beta2
alpha = model.params['const']
beta1 = model.params['age']
beta2 = model.params['BMI']

print("A1", alpha)
print("Beta1 ", beta1)
print("Beta2 ", beta2)
```
