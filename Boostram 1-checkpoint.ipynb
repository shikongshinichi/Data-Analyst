```python
import os, shutil
```

```python
path = r"C:/Users/ADMIN/Pictures/"
```

```python
filename = os.listdir(path)
```

```python
os.path.exists(path + "csv files")
```

```python
folname = ['csv file', 'image file', 'text file']
for loop in range(0, 3):
    if not os.path.exists(path + folname[loop]):
        print(path + folname[loop])
        os.makedirs(path + folname[loop])
```

```python
for file in filename:
    if ".csv" in file and not os.path.exists(path + "csv file/" + file):
        shutil.move(path + file, path + "csv file/" + file)
    elif ".png" in file and not os.path.exists(path + "image file/" + file):
        shutil.move(path + file, path + "image file/" + file)
    else:
        print("Not file is moved!")
```
