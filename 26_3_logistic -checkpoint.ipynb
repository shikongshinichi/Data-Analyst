```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.info()
```

```python
#tạo bảng sao
df_copy = df.copy()
#lưu vào ổ đĩa
df_copy.to_csv('titanic.csv')
```

# mô hình 1 là xóa các giá trị missing values

```python
#kiểm tra số lượng tổng các missing value
df.isna().sum()
```

```python
df = df.dropna() #xoá các giá trị missing value
df.isnull().sum()# kiểm tra lại
```

```python
df = df[['Survived', 'Age', 'Sex', 'Pclass']]
df = pd.get_dummies(df, columns=['Sex', 'Pclass'])
df.dropna(inplace=True)
df.head()
```

```python

from sklearn.model_selection import train_test_split

x = df.drop('Survived', axis=1)
y = df['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, x, y, cv=5).mean()
```

```python
from sklearn.metrics import confusion_matrix

y_predicted = model.predict(x_test)
confusion_matrix(y_test, y_predicted)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Tính ma trận nhầm lẫn
cm = confusion_matrix(y_test, model.predict(x_test))

# Hiển thị ma trận nhầm lẫn bằng matplotlib
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
plt.xticks([0, 1], ['Perished', 'Survived'])
plt.yticks([0, 1], ['Perished', 'Survived'])
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predicted))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Tính các giá trị cho đường cong ROC
fpr, tpr, thresholds = roc_curve(y_test, model.predict_proba(x_test)[:,1])

# Tính diện tích dưới đường cong ROC
roc_auc = auc(fpr, tpr)

# Vẽ đường cong ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

# mô hình 2 không xóa mà thay thế các giá trị missing value bằng giá trị phù hợp (biến số thay thế = mean; biến chuổi thay thế =mode (giá trị xuất hiện nhiều nhất)

```python
# Tính mean của cột 'Age'
mean_age = df_copy['Age'].mean()

# Thay thế các giá trị thiếu trong cột 'Age' bằng mean
df_copy['Age'].fillna(mean_age, inplace=True)

# Tính mode của cột 'Embarked'
mode_embarked = df_copy['Embarked'].mode()[0]

# Thay thế các giá trị thiếu trong cột 'Embarked' bằng mode
df_copy['Embarked'].fillna(mode_embarked, inplace=True)
```

```python
df_copy = df_copy[['Survived', 'Age', 'Sex', 'Pclass']]
df_copy = pd.get_dummies(df_copy, columns=['Sex', 'Pclass'])
df_copy.dropna(inplace=True)
df_copy.head()
```

```python
from sklearn.model_selection import train_test_split

x = df_copy.drop('Survived', axis=1)
y = df_copy['Survived']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(model, x, y, cv=5).mean()
```

```python
from sklearn.metrics import confusion_matrix

y_predicted = model.predict(x_test)
confusion_matrix(y_test, y_predicted)
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Load iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Train a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title('Decision Tree')
plt.show()
```

```python
# Plot the decision tree
plt.figure(figsize=(10, 8))
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
